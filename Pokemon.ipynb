```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def createData(shape):
    data = np.zeros(shape)
    data.fill(1.0)
    
    return data

def drawLine(x1, y1, R, G, B, data):
    
    x = x1
    y = y1
    x2 = x1 
    y2 = y1 +10
    
    if x1 < x2:
        xi = 1
        dx = x2 - x1
    else:
        xi = -1
        dx = x1 - x2
        
    if y1 < y2:
        yi = 1
        dy = y2-y
    else:
        yi = -1
        dy = y1 - y2
            
    
    if dx > dy:
        ai = (dy - dx) * 2
        bi = dy * 2
        d = bi - dx
        while x != x2:
            if d >= 0:
                x += x1
                y += yi
                d += ai
            else:
                d += bi
                x += xi
    
            for i in range(11):
                data[x+i,y,0] = R
                data[x+i,y,1] = G
                data[x+i,y,2] = B
    else:
        ai = (dx - dy) * 2
        bi = dx * 2
        d = bi - dy
        while y != y2:
            if d >= 0:
                x += xi
                y += yi
                d += ai
            else:
                d += bi
                y += yi
            for i in range(11):
                data[x+i,y,0] = R
                data[x+i,y,1] = G
                data[x+i,y,2] = B
                
                
                
             
def drawRectangle(x, y, R, G, B, data):
    drawLine(x, y, R, G ,B, data)
    
def drawYellow(x, y, d):
    drawRectangle(x, y, 1, 1, 0, d)
    
def drawRed(x, y, d):
    drawRectangle(x, y, 1, 0, 0, d)
    
def drawBlack(x, y, d):
    drawRectangle(x, y, 0, 0, 0, d)
    
def drawOrange(x, y, d):
    drawRectangle(x, y, 1, 0.647, 0, d)
```
